```python
pages = []

#geenrate the urls through the iterations
for i in iterations:
    start_at = str(i)
    url = f'https://www.emsc-csem.org/Earthquake/?view={i}'
    
    #download the url
    response = requests.get(url)
    pages.append(response)
pages

# now the connections are created
```

```python
#soup_pages = BeautifulSoup(pages[0].content, 'html.parser')

for i in range(5):
    pages_soup=BeautifulSoup(pages[i].content, 'html.parser')
pages_soup
```

```python
titles=[pages_soup.select(f'#reg{i}') for i in range(51)]

len(titles)
```

```python
titles2 = pages_soup.select('.tb_region')
len(titles2)
```

```python
names = [nam.text for name in titles for nam in name]

names
```

```python
cities = [name.lstrip('\xa0') for name in names]
len(cities)
```
