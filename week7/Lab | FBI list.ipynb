```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
from tqdm.notebook import tqdm
```

```python
url = 'https://www.fbi.gov/wanted/topten'
response = requests.get(url)
html= response.content
scraped = soup(html, "html.parser")
```

```python
scraped
```

```python
wanted_people = scraped.select("li.portal-type-person > h3 > a")
wanted_people
```

```python
wanted = []
for man in wanted_people:
    wanted.append(man.text)
wanted
```
