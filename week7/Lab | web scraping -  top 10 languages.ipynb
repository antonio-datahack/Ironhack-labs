## Assignment

Display the top 10 languages by number of native speakers stored in a pandas dataframe: url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
from tqdm.notebook import tqdm
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers"
response = requests.get(url)
html= response.content
scraped = soup(html, "html.parser")
```

```python
list_of_languages = scraped.select("table.wikitable > tbody > tr > td > a")
```

```python
lista = [language.text for language in list_of_languages]
```

```python
lista
```

```python
speakers1 = scraped.select("table.wikitable > tbody > tr > td")
```

```python
speakers =[speaker.text for speaker in speakers1]
```

```python
count = 0
rank= []
language = []
speakerss = []
percentage = []
language_family = []
branch = []

while count < 55:
    rank.append(speakers[count][:-1])
    count +=1
    language.append(speakers[count][:-1])
    count +=1
    speakerss.append(speakers[count][:-1])
    count +=1
    percentage.append(speakers[count][:-1])
    count +=1
    language_family.append(speakers[count][:-1])
    count +=1
    branch.append(speakers[count][:-1])
    count +=1
```

```python
df = pd.DataFrame({"rank" : rank,
                    "language": language,
                    "speakers": speakerss,           
                    "percentage": percentage,
                    "language_family": language_family,
                    "branch":branch
});
```
